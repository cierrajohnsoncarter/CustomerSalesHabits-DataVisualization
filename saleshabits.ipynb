Import the libraries

```
import matplotlib.pyplot as plt
```

```
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]
```

```
# Numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]
```

Create a figure

```
plt.figure(figsize=(24, 16))
```

Create a subplot of x values vs total page visits over the past year as a line

```
ax1 = plt.subplot(1, 2, 1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker='o')

plt.xlabel('Months')
plt.ylabel('Visits per Month')
plt.title('Visits Over 12 Months')

ax1.set_xticks(x_values)
ax1.set_xticklabels(months, rotation=90)
```

Create a subplot to represent the # of key limes sold, # of persian limes sold, and # of blood limes sold vs x values

```
ax2 = plt.subplot(1, 2, 2)
plt.plot(x_values, key_limes_per_month, marker='o', color='blue')
plt.plot(x_values, persian_limes_per_month, marker='o', color='red')
plt.plot(x_values, blood_limes_per_month, marker='o', color='green')

ax2.set_xticks(x_values)
ax2.set_xticklabels(months, rotation=90)

plt.title('Limes Sold per Month')
plt.legend(['key limes', 'persian limes', 'blood limes'])
plt.show()
```
